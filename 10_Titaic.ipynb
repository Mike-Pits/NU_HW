```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
train_data.index = train_data.PassengerId
train_data
```

```python
y = train_data['Survived'].values
```

```python
train_data.head()
```

```python
del train_data['PassengerId']
train_data.head()
```

```python
del train_data['Survived']
train_data.head()
```

```python
train_data['Has_Cabin'] = train_data['Cabin'].apply(lambda x: 0 if type(x)==float else 1)
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
train_data['Is_alone'] = train_data['FamilySize'].apply(lambda x: 0 if x > 1 else 1)
```

```python
train_data.head()
```

```python
import re

def get_title(name):
    title_search = re.search('([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ''

train_data['Title'] = train_data['Name'].apply(get_title)
train_data.head(3)
```

```python
train_data['Title'].value_counts()
```

```python
rare = ['Dr', 'Rev', 'Major', 'Col', 'Countess', 'Capt', 'Lady', 'Don', 'Jonkheer']
train_data['Title'] = train_data['Title'].replace(rare, 'Rare')
train_data['Title'] = train_data['Title'].replace('Mlle', 'Miss')
train_data['Title'] = train_data['Title'].replace('Ms', 'Miss')
train_data['Title'] = train_data['Title'].replace('Mme', 'Mrs')
train_data['Title'] = train_data['Title'].replace('Sir', 'Mr')
```

```python
train_data['Title'].value_counts()
```

```python
train_data = train_data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
train_data.head(3)
```

```python
title_mapping = {'Mr': 1, 'Master': 2., 'Mrs': 3, 'Miss': 4, "Rare": 5}
train_data['Title'] = train_data['Title'].map(title_mapping)
```

```python
train_data['Title'].isna().value_counts()
```

```python
len(train_data)
```

```python
map_sex = {'male': 1, 'female': 0}
train_data['Sex'] = train_data['Sex'].map(map_sex)
```

```python
train_data['Sex'].isna().value_counts()
```

```python
map_embarked = {'S': 0, 'C': 1, 'Q': 2}
train_data['Embarked'] = train_data['Embarked'].map(map_embarked)
train_data['Embarked'].isna().value_counts()
```

```python
train_data.head(3)
```

```python
train_data.info()
```

```python
train_data['Embarked'] = train_data['Embarked'].fillna(value=0)
train_data['Age'] = train_data['Age'].fillna(value=train_data['Age'].mean()) 
```

```python
train_data.info()
```

```python
train_data.head(1)
```

```python
cat_cols = ['Pclass', 'Sex', 'SibSp', 
            'Parch', 'Embarked',
           'Has_Cabin', 'FamilySize', 'Is_alone', 'Title']
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)
encoded_cat_cols = pd.DataFrame(onehot_encoder.fit_transform(train_data[cat_cols]),
                               index=train_data.index)
encoded_cat_cols.head(10)
```

```python
res_data_all = train_data.join(encoded_cat_cols)
res_data_all.head(3)
```

```python
res_data_all = res_data_all.drop(cat_cols, axis=1)
res_data_all.head(3)
```

```python
from sklearn.preprocessing import StandardScaler

nn_scaler = StandardScaler()
res_all_data_scaled = nn_scaler.fit_transform(res_data_all)
```

```python
res_all_data_scaled.shape
```

```python
res_all_data_scaled[0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(res_all_data_scaled,
                                                 y,
                                                 test_size=0.10,
                                                 random_state=42)
```
